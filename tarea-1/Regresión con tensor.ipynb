<h5>funci√≥n para entrenar un modelo de regresi√≥n lineal y = mx +b.</h5>

```python
def lineal_regression(x,y,epochs,lr,print_interval):
    
#     tf.reset_default_graph()
    g = tf.Graph()
    
    parametros = tf.zeros((2,1), tf.float64)
                          
    with tf.name_scope("parametros_w_b"):
        weights = tf.Variable(name="w", initial_value = parametros)
        x = tf.concat([x, tf.ones_like(x)], axis=1)
    with tf.name_scope("Hipotesis"):
        y_hat = tf.matmul(x, weights)
    with tf.name_scope("Error"):
        error = (0.5 * tf.reduce_mean(tf.math.square(y - y_hat)))
    with tf.name_scope("Gradientes"):
        gradients = tf.gradients(error, weights)
    with tf.name_scope("Update_Weights"):
        weights_update = tf.compat.v1.assign(weights, (tf.add(weights, tf.scalar_mul(-lr,gradients[0]))))
                          
    error_summary = tf.compat.v1.summary.scalar(name="Error_vs_Epochs", tensor=error)
    with tf.compat.v1.Session() as session:
        curr_datetime = datetime.datetime.now().strftime('%Y%m%d_%H%M%S')
        writer = tf.compat.v1.summary.FileWriter(".tensorboard/" + curr_datetime + "_lr=" + str(lr) + "_epochs=" + str(epochs), session.graph)
        session.run(tf.compat.v1.global_variables_initializer())
        
        for epoch in range(0, epochs):
            if ((epochs + 1) % print_interval) == 0:
                print("epoch:" + str(epoch + 1) + " error: " + str(session.run(error)))
            
            writer.add_summary(session.run(error_summary), epoch)
            session.run(weights_update)
                          
                          
                          
                          
                          
        writer.close()
    session.close()
                          
        
```

```python
dataset  = np.load('data/proyecto_training_data.npy')

dataset[0:,0]/=1000

saleprice = dataset[:,0].reshape(-1,1)
overallquall = dataset[:,0].reshape(-1,1)
```

```python
lineal_regression(overallquall,saleprice, 10000, 0.001, 200)
```
